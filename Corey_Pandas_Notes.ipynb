```python
# **Part 4: Filtering - Using conditionals to filter rows and columns**

# Using the alternative df

people = {
    "first": ["Corey", 'jane', 'John'],
    "last": ['Schafer', 'Doe', 'Doe'],
    "email": ['CoreyMSchafer@gmail.com', 'janeDoe@email.com', 'JohnDoe@email.com']
}
import pandas as pd 
df = pd.DataFrame(people)
df
```

```python
# Define a variable filt and filter all the last names with Schafer and first names John
filt = (df["last"] == "Schafer") | (df["first" ] == "John")
```

```python
df.loc[filt, "email"]
# As we specified that we want the email column, we get the email column that apply to the filter
```

```python
df.loc[-filt, "email"]
# We get the emails that are not affected by the filter.
```
